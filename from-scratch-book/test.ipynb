```python
import torch

sizes = [10, 100, 512, 1024, 2048, 4096]

for n in sizes:
    print(f"\nMatrix size: {n} x {n}")

    a_cpu = torch.randn(n, n)
    b_cpu = torch.randn(n, n)

    a_gpu = a_cpu.to("cuda")
    b_gpu = b_cpu.to("cuda")

    print("CPU:")
    %timeit a_cpu @ b_cpu

    print("GPU:")
    %timeit torch.cuda.synchronize(); a_gpu @ b_gpu; torch.cuda.synchronize()
```
